```python
import pandas as pd


df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
nan_matrix = df.isnull()
nan_matrix.sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mode()[0])
df["Cabin"] = df["Cabin"].fillna("Unknown")
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
nan_matrix = df.isnull()
nan_matrix.sum()
```

```python
from sklearn.preprocessing import StandardScaler

a = ["Age", "Fare", "SibSp", "Parch"]
df[a] = StandardScaler().fit_transform(df[a])

print(df[a].describe())
```

```python
c_c = df.select_dtypes(include = ['object']).columns

one_cols = ["Sex", "Embarked", "Cabin"]
drop = ["Name", "Ticket"]

df["Cabin"] = df["Cabin"].str[0]
df["Cabin"] = df["Cabin"].fillna('Unknown')

df = pd.get_dummies(df, columns = one_cols , drop_first = True, prefix = one_cols)

df.drop(drop, axis = 1, inplace = True, errors = 'ignore')

df.head()
```

```python
df.to_csv("processed_titanic.csv", index = False)
```
